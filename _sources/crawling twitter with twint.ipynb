## 2. Import Library Twint

```python
import twint
import pandas
import nest_asyncio
nest_asyncio.apply()
```

## 3. Crawling data twitter dengan Twint

```python
config = twint.Config()
config.Search = "Cak Imin"
config.Lang = "id"
config.Limit = 500
config.Since = "2022-01-01"
config.Until = "2023-02-16"
config.Pandas = True

twint.run.Search(config)
```

* DataFrame hasil crawling twitter

```python
data = twint.storage.panda.Tweets_df
df = pandas.DataFrame(data, columns=['date', 'username', 'tweet'])
df
```
